```python
area1 = '부산'
area2 = '중구'
month1 = 5
day1 = 6
hour1 = 6
month2 = 6
day2 = 2
hour2 = 18
```

```python
start_date = datetime(2024, month1, day1, hour1)
end_date = datetime(2024, month2, day2, hour2)

print(start_date)
print(end_date)
```

```python
co_area1 = area1 + ' ' + area2
```

```python
query = """
        SELECT ds, yhat
        FROM pred
        WHERE area = :co_area1
            AND ds BETWEEN TO_TIMESTAMP(:start_date, 'YYYY-MM-DD HH24:MI:SS')
                       AND TO_TIMESTAMP(:end_date, 'YYYY-MM-DD HH24:MI:SS')
        order by ds
        """
```

```python
# 바인딩된 변수 설정
bind_params = {
    'area': co_area1,
    'start_date': start_date.strftime('%Y-%m-%d %H:%M:%S'),
    'end_date': end_date.strftime('%Y-%m-%d %H:%M:%S')
}
```

```python
import pandas as pd 
db.init_oracle_client()
engine = create_engine('oracle+oracledb://jennie:12345@localhost:1521/?service_name=xe')
pred_df = pd.read_sql(query, con=engine, params=bind_params)
print(pred_df)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# 기준선 설정
thresholds = [30, 90, 150]

# 기준선에 따라 데이터 구간 나누기
threshold1 = pred_df[pred_df['yhat'] <= thresholds[0]]
threshold2 = pred_df[(pred_df['yhat'] > thresholds[0]) & (pred_df['yhat'] <= thresholds[1])]
threshold3 = pred_df[(pred_df['yhat'] > thresholds[1]) & (pred_df['yhat'] <= thresholds[2])]
threshold4 = pred_df[pred_df['yhat'] > thresholds[2]]

# 데이터 구간별로 그래프 그리기
plt.plot(threshold1['ds'], threshold1['yhat'], linestyle='-', color='b', label=f'≤ {thresholds[0]} ㎍/m³')
plt.plot(threshold2['ds'], threshold2['yhat'], linestyle='-', color='g', label=f'{thresholds[0]} - {thresholds[1]} ㎍/m³')
plt.plot(threshold3['ds'], threshold3['yhat'], linestyle='-', color='y', label=f'{thresholds[1]} - {thresholds[2]} ㎍/m³')
plt.plot(threshold4['ds'], threshold4['yhat'], linestyle='-', color='r', label=f'> {thresholds[2]} ㎍/m³')

# 기준선 추가
for threshold in thresholds:
    plt.axhline(y=threshold, color='b', linestyle='--', label=f'기준선 {threshold} ㎍/m³')

plt.xlabel('날짜')
plt.ylabel('미세먼지 농도(PM10)')
plt.title(f'{co_area1} 미세먼지 농도(PM10) 예측치(단위: ㎍/m³)')
plt.xticks(rotation=45)  # x축 레이블 회전

# X축 날짜 형식 설정
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))  # 날짜 간격을 1일로 설정
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # 날짜 형식을 '연도-월-일'로 설정

plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# 데이터 필터링
filtered_df = pred_df[(pred_df['ds'] >= start_date) & (pred_df['ds'] <= end_date)]

# 그래프 그리기
plt.figure(figsize=(10, 5))

# yhat 값에 따라 색상 변경, 기준선 추가
for i in range(1, len(filtered_df)):
    if filtered_df['yhat'].iloc[i] <= 30:
        plt.plot(filtered_df['ds'].iloc[i-1:i+1], filtered_df['yhat'].iloc[i-1:i+1], color='b')
        plt.axhspan(min(filtered_df['yhat']), 30, facecolor='lightblue', alpha=0.5)
    elif filtered_df['yhat'].iloc[i] > 30:
        plt.plot(filtered_df['ds'].iloc[i-1:i+1], filtered_df['yhat'].iloc[i-1:i+1], color='g')
        plt.axhspan(30, max(filtered_df['yhat']), facecolor='lightgreen', alpha=0.5)
        plt.axhline(y=30, color='r', linestyle='--', linewidth=1)
    elif filtered_df['yhat'].iloc[i] > 90:
        plt.plot(filtered_df['ds'].iloc[i-1:i+1], filtered_df['yhat'].iloc[i-1:i+1], color='y')
        plt.axhline(y=90, color='r', linestyle='--', linewidth=1)
        plt.axhspan(90, max(filtered_df['yhat']), facecolor='lightyellow', alpha=0.5)
    else :
        plt.plot(filtered_df['ds'].iloc[i-1:i+1], filtered_df['yhat'].iloc[i-1:i+1], color='r')
        plt.axhline(150, max(filtered_df['yhat']), color='r', linestyle='--', linewidth=1)
        
plt.xlabel('날짜')
plt.ylabel('미세먼지 농도(PM10)')
plt.title(f'{co_area1} 미세먼지 농도(PM10) 예측치(단위: ㎍/m³)')
plt.xticks(rotation=45)  # x축 레이블 회전
plt.grid(True)
plt.legend()
plt.tight_layout()

# X축 날짜 형식 설정
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))  # 날짜 간격을 1일로 설정
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # 날짜 형식을 '연도-월-일'로 설정

plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
